```python
keywords_1 = ['PT', 'Depdagri', 'DITJEN', 'kemenkeu', 'depkeu', 'sd', 'smp', 'sma', 'kementerian', 'kementrian', 'kppn', 'kpp', 'kemendagri', 'polkum', 'kabinet', 'sekretariat', 'negara', 'astra', 'dephub', 'ppatk', 'ojk', 'lembaga', 'departemen', 'unand', 'tk', 'direktorat', 'ykk', 'mahkamah', 'bank', 'dirjen', 'sriwijaya', 'pgn', 'kemenkes', 'dpr', 'brin', 'kimia farma', 'sekneg', 'setjen', 'airnav', 'djp', 'bmkg', 'basarnas', 'sar', 'kejaksaan', 'rs', 'rumah sakit', 'badan', 'askrindo', 'bazarnas', 'airlines', 'air asia', 'komisi', 'walikota', 'tripatra', 'lpmi', 'tni', 'tentara', 'polisi', 'polri', 'indomarco', 'lemhanas', 'dinas', 'telekom', 'telkom', 'arsip', 'bps', 'otoritas', 'pos', 'bea cukai', 'pemda', 'pgas', 'kelurahan', 'kecamatan']
keywords_1 = [keyword.lower() for keyword in keywords_1]
keywords_2 = ['Pemilik', 'Manager', 'manajer', 'OWNER', 'Kontrakan', 'Kosan']
keywords_2 = [keyword.lower() for keyword in keywords_2]
data_fix['segment_company_pt'] = np.where(data_fix['employer_company_name'].str.contains('|'.join(keywords_1)), 1, 0)
data_fix['segment_company_owner'] = np.where(data_fix['employer_company_name'].str.contains('|'.join(keywords_2)), 1, 0)

data_fix
```

```python
data_fix[data_fix['segment_company_owner']==1]
```

```python
data_fix[data_fix['employer_company_name']=='perseorangan ibu rumah tangga']
```

```python
data_fix['segment_final'] = ''

for index, row in data_fix.iterrows():
    if row['segment_company_pt'] == 1 and row['segment_company_owner'] == 0:
        data_fix.at[index, 'segment_final'] = 1
    elif row['segment_company_pt'] == 0 and row['segment_company_owner'] == 1:
        data_fix.at[index, 'segment_final'] = 2
    elif row['segment_company_pt'] == 0 and row['segment_company_owner'] == 0:
        data_fix.at[index, 'segment_final'] = 3
    else:
        data_fix.at[index, 'segment_final'] = 2
```

```python
data_fix
```

```python
data_fix['indicative_limit'].max()
```

```python
data_fix['flag'] = ''

for index, row in data_fix.iterrows():
    if row['segment_final'] == 3:
        data_fix.at[index, 'flag'] = 1
    elif row['segment_final'] == 2 and row['indicative_limit'] < 30000000:
        data_fix.at[index, 'flag'] = 2
    elif row['segment_final'] == 2 and row['indicative_limit'] >= 30000000:
        data_fix.at[index, 'flag'] = 3
    elif row['segment_final'] == 1:
        data_fix.at[index, 'flag'] = 4
data_fix
```

```python
data_fix['flag'] = data_fix['flag'].astype('int32')
data_fix['segment_final'] = data_fix['segment_final'].astype('int32')
data_fix.dtypes
```

```python
X = data_fix[['segment_final','indicative_limit']]
X
```

```python
y = data_fix['flag']
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 123)
```

```python
# Create function to evaluate our model
def classification_eval (aktual, prediksi, name):
    cm = confusion_matrix(aktual, prediksi)
    tp = cm[1][1]
    tn = cm[0][0]
    fp = cm[0][1]
    fn = cm[1][0]
    
    accuracy = round((tp+tn) / (tp+tn+fp+fn) * 100, 2)
    precision = round((tp) / (tp+fp) * 100, 2)
    recall = round((tp) / (tp+fn) * 100, 2)
    specificity = round((tn) / (tn+fp) * 100, 2)
    
    print('Evaluation Model:', name)
    print(cm)
    print('Accuracy   :', accuracy, '%')
    print('Precision  :', precision, '%')
    print('Recall     :', recall, '%')
    print('Specificity:', specificity, '%')
```

```python
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
y_train_pred = knn.predict(X_train)
y_test_pred = knn.predict(X_test)
classification_eval(y_train, y_train_pred, 'KNN Training Perf.')
```

```python
classification_eval(y_test, y_test_pred, 'KNN Testing Perf.')
```

```python
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
y_train_pred = dt.predict(X_train)
y_test_pred = dt.predict(X_test)
classification_eval(y_train, y_train_pred, 'Dectree Training Perf.')
```

```python
classification_eval(y_test, y_test_pred, 'Dectree Testing Perf.')
```

```python
rf = RandomForestClassifier(random_state = 123)
rf.fit(X_train, y_train)
y_train_pred = rf.predict(X_train)
y_test_pred = rf.predict(X_test)
classification_eval(y_train, y_train_pred, 'Ranfor Training Perf.')
```

```python
classification_eval(y_test, y_test_pred, 'Ranfor Testing Perf.')
```
